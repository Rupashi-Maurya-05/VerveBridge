Data Cleaning

```python
# Selected Players
df1=df
df=df.dropna(subset=['Draft pick'])

# Players not selected or their draft pick data is missing
nan_df=df1[df1['Draft pick'].isna()]
```

```python
# Dropping rows with all null values
nan_df = nan_df.dropna(axis=1, how='all')
```

```python
# nan_df after cleaning
nan_df.info()
```

```python
# Handling missing values

# Dropping columns with more than 40% missing values
df = df.dropna(thresh=len(df) * 0.6, axis=1)
```

```python
# Dataset after removing column 'Bench' whith more than 40% missing values
df.head()
```

```python
#removing 'Bench' column from nan_df
nan_df=nan_df.drop(columns=['Bench'])
```

```python
nan_df.head()
```

```python
# Dropping column we don't need for analysis
df = df.drop(columns=['Unnamed: 0'])
nan_df = nan_df.drop(columns=['Unnamed: 0'])
```

```python
# Outlier reduction using the IQR method
def reduce_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Select only numeric columns
numeric_columns = df.select_dtypes(include=[np.number]).columns
for column in numeric_columns:
    reduce_outliers(df, column)
```

```python
# Filling missing values in numeric columns with the median of each column
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].median())

# Dataset after cleaning
df.info()
```

```python
# Filling missing values in numeric columns with the median of each column in data of players with draft value as nan
numeric_columns = nan_df.select_dtypes(include=[np.number]).columns
nan_df[numeric_columns] = nan_df[numeric_columns].fillna(nan_df[numeric_columns].median())

# Non Selected Players Dataset after cleaning
nan_df.info()
```

Trend Analysis Over Time:
Visualising Yearly Trends

```python
# Selected Players
# Ensuring that the columns to be averaged are numeric
df.loc[:, ['Year', 'Draft pick', 'Height (No Shoes)',
       'Height (With Shoes)', 'Wingspan', 'Standing reach', 'Vertical (Max)',
       'Vertical (Max Reach)', 'Vertical (No Step)',
       'Vertical (No Step Reach)', 'Weight', 'Body Fat', 'Hand (Length)',
       'Hand (Width)', 'Agility', 'Sprint']] = df[['Year', 'Draft pick', 'Height (No Shoes)',
       'Height (With Shoes)', 'Wingspan', 'Standing reach', 'Vertical (Max)',
       'Vertical (Max Reach)', 'Vertical (No Step)',
       'Vertical (No Step Reach)', 'Weight', 'Body Fat', 'Hand (Length)',
       'Hand (Width)', 'Agility', 'Sprint']].apply(pd.to_numeric, errors='coerce')
df_clean = df[['Year', 'Draft pick', 'Height (No Shoes)',
       'Height (With Shoes)', 'Wingspan', 'Standing reach', 'Vertical (Max)',
       'Vertical (Max Reach)', 'Vertical (No Step)',
       'Vertical (No Step Reach)', 'Weight', 'Body Fat', 'Hand (Length)',
       'Hand (Width)', 'Agility', 'Sprint']].dropna()

yearly_avg = df_clean.groupby('Year').mean()

# Year-wise comparison using Line plot for selected players
plt.figure(figsize=(14, 8))
plt.plot(yearly_avg, marker='o')
plt.title('Year-wise Average Physical Attributes (2009-2017) of selected players')
plt.xlabel('Year')
plt.ylabel('Average Value')
plt.legend(yearly_avg.columns, loc='upper right')
plt.grid(True)
plt.show()
```

```python
# Players that Were Not Selected
# Select only the numeric columns for grouping and averaging
numeric_columns = ['Year', 'Height (No Shoes)', 'Height (With Shoes)', 'Wingspan',
                   'Standing reach', 'Vertical (Max)', 'Vertical (Max Reach)',
                   'Vertical (No Step)', 'Vertical (No Step Reach)', 'Weight',
                   'Body Fat', 'Hand (Length)', 'Hand (Width)', 'Agility', 'Sprint']

# Ensure that only numeric columns are passed to groupby
nan_df_clean = nan_df[numeric_columns].dropna()

# Group by 'Year' and compute the mean for numeric columns
yearly_avg = nan_df_clean.groupby('Year').mean()

# Year-wise comparison using Line plot
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 8))
for column in yearly_avg.columns:
    plt.plot(yearly_avg.index, yearly_avg[column], marker='o', label=column)

plt.title('Year-wise Average Physical Attributes (2009-2017) of players that were not selected')
plt.xlabel('Year')
plt.ylabel('Average Value')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

```python
# Heatmap of average physical attributes over years
plt.figure(figsize=(12, 10))
sns.heatmap(yearly_avg.T, cmap='viridis', annot=True, fmt='.1f', linewidths=.5)
plt.title('Heatmap of Average Physical Attributes by Year (2009-2017)')
plt.xlabel('Year')
plt.ylabel('Physical Attributes')
plt.xticks(rotation=45)  
plt.tight_layout()  
plt.show()
```

Visualising Meaningful Relationships between Attributes

```python
# Converting relevant columns to numeric, coercing errors to NaN
columns_to_convert = ['Height (With Shoes)', 'Wingspan', 'Vertical (Max)', 'Weight', 'Draft pick', 'Agility', 'Sprint', 'Body Fat']
df.loc[:, columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')
df_clean = df.dropna(subset=columns_to_convert)
```

```python
# Analyzing relation between Physical Attributes vs Draft Pick using Regression Plot 

fig, axes = plt.subplots(3, 2, figsize=(14, 18))
fig.suptitle('Physical Attributes vs. Draft Pick', fontsize=16)

# Height vs. Draft Pick
sns.regplot(x='Height (With Shoes)', y='Draft pick', data=df_clean, ax=axes[0, 0], scatter_kws={'color': 'blue'}, line_kws={'color': 'blue'})
axes[0, 0].set_title('Height (With Shoes) vs. Draft Pick')
axes[0, 0].invert_yaxis()  # Lower draft picks (higher number) are better

# Wingspan vs. Draft Pick 
sns.regplot(x='Wingspan', y='Draft pick', data=df_clean, ax=axes[0, 1], scatter_kws={'color': 'green'}, line_kws={'color': 'green'})
axes[0, 1].set_title('Wingspan vs. Draft Pick')
axes[0, 1].invert_yaxis()

# Vertical Jump vs. Draft Pick 
sns.regplot(x='Vertical (Max)', y='Draft pick', data=df_clean, ax=axes[1, 0], scatter_kws={'color': 'red'}, line_kws={'color': 'red'})
axes[1, 0].set_title('Vertical Jump vs. Draft Pick')
axes[1, 0].invert_yaxis()

# Agility vs. Draft Pick 
sns.regplot(x='Agility', y='Draft pick', data=df_clean, ax=axes[1, 1], scatter_kws={'color': 'magenta'}, line_kws={'color': 'purple'})
axes[1, 1].set_title('Agility vs. Draft Pick')
axes[1, 1].invert_yaxis()

# Sprint vs. Draft Pick
sns.regplot(x='Sprint', y='Draft pick', data=df_clean, ax=axes[2, 0], scatter_kws={'color': 'orange'}, line_kws={'color': 'orange'})
axes[2, 0].set_title('Sprint vs. Draft Pick')
axes[2, 0].invert_yaxis()

# Body Fat Percentage vs. Draft Pick 
sns.regplot(x='Body Fat', y='Draft pick', data=df_clean, ax=axes[2, 1], scatter_kws={'color': 'cyan'}, line_kws={'color': 'cyan'})
axes[2, 1].set_title('Body Fat Percentage vs. Draft Pick')
axes[2, 1].invert_yaxis()

plt.tight_layout(rect=[0, 0, 1, 0.96])  
plt.show()
```
